# Desafio.

1. Realizar a tokenização de palavras do artigo
2. Realizar a distribuição de frequência das palavras (liberado geral o uso de novos stopwords) - FreqDist e grafico de linha.
3. Criar o wordcloud;
4. Apresentar e receber o certificado

```python
texto=artigo.cleaned_text
```

```python
texto
```

```python
type(texto)
```

```python
from nltk.tokenize import word_tokenize
```

```python
palavras=word_tokenize(texto)
```

```python
palavras
```

```python
from nltk.probability import FreqDist
```

```python
freq=FreqDist(palavras)
```

```python
freq
```

```python
freq.most_common(10)
```

```python
freq.plot(10)
```

```python
freq.plot(10,cumulative=True)
```

```python
freq.plot(20)
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
! pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
# Definição da Função
def plot_cloud(wordcloud):
    # Criar a Figura
    plt.figure(figsize=(40, 30))

    # Exibir a Word Cloud
    plt.imshow(wordcloud)

    # Remover Eixos
    plt.axis("off")
```

```python
wordcloud = WordCloud(width = 3000, height = 2000, random_state = 1,
background_color='salmon', colormap='Pastel1', collocations=False,
stopwords=STOPWORDS).generate(texto)
```

```python
plot_cloud(wordcloud)
```
