# Inferir especie de uma flor a partir das medidas destas

### List Comprehension

```python
df.columns=[s.replace('.','_') for s in df.columns ]

df.head()
```

```python
df['variety'].value_counts()
```

## obs: Estão bem distribuidas
## E as meidas?

```python
df_setosa=df.query('variety == "Setosa"').copy()
df_versicolor=df.query('variety== "Versicolor"').copy()
df_virginica = df.query('variety == "Virginica"').copy()
```

```python
medidas = ['sepal_length','sepal_width','petal_length','petal_width']
```

```python
! pip install matplotlib
```

```python
print('Setosa')
df_setosa[medidas].hist()
```

```python
print('Versicolor')
df_versicolor[medidas].hist()
```

```python
print('Virginica')
df_virginica[medidas].hist()
```

```python
df.query('petal_length >5')['variety'].value_counts()
```

# Conclusão: Praticamente só as plantas da especie Virginica tem comprimento de pétala maior que 5!

- Nenhuma amostra da setosa posssui essas medidas maior que 5.

* Isso nos diz que a variavel "petal_lenght" possuir um bom pode preditivo referente ao tipo de iris
* Quais amostras são de uma espécie e quais não são
