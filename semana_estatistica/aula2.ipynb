# Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Importando o dataset

```python
df=pd.read_csv('datasets/imigrantes_canada.csv')
```

```python
df.head()
```

```python
type(df)
```

```python
df.info()
```

```python
df.columns
```

## Realizando alguns tratamentos de dados

## Brasil

### Atribuindo o index a país

```python
df.set_index('País',inplace=True)
```

```python
df.head()
```

### Filtrando quantidade de anos de 1980 a 2013

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
type(anos)
```

# Extrair a série de dados do Brasil (loc[])

- loc []- pesquisa pelo label
- iloc []- pesquisa pelo indice

```python
# extrair a serie de dados do brasil

df_brasil=df.loc['Brasil',anos]
df_brasil
```

```python
type(df_brasil)
```

```python
df_brasil.plot()
```

```python
df_brasil_dict= {'ano':df_brasil.index.tolist(),'imigrantes':df_brasil.values.tolist()}
```

```python
type(df_brasil_dict)
```

```python
df_brasil_dict
```

```python
df_brasil=pd.DataFrame(df_brasil_dict)
```

```python
df_brasil.info()
```

```python
df_brasil.describe()
```

```python
df_brasil.head(100)
```

# Criando o grafico de linhas

```python
df_brasil.plot(kind='line')
```

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.grid()
plt.show()
```

# Transformando o grafico em figura
* Vantagens
    * Aumenta o potencial dos comandos
    

```python
plt.figure (figsize=(8,4))

# Altura - 8 polegadas

# Largura - 4 polegadas

# 1 polegada = 2.54 cm

# criar uma figura

plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canada')
plt.xlabel('Ano')
plt.ylabel('Numero de imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.grid()
plt.show()

```

# Criando subplots
    * Varios graficos juntos em um grafico só

```python
fig , ax = plt.subplots(figsize=(8,4))

ax.plot(df_brasil['ano'],df_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canada \n 1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Numero de imigrantes')
ax.grid()
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # insere um intervalo no eixo x de 5 e0 5 no caso
ax.yaxis.set_major_locator(plt.MultipleLocator(300)) # insere um intervalo no eixo y de 1000 em 1000 no caso
plt.show()
```

# Gráfico de linhas e boxplot em subplots


```python
fig, axs = plt.subplots(1,2,figsize=(10,4))

# subplot 1

axs[0].plot(df_brasil['ano'],df_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canada \n 1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Numero de imigrantes')
axs[0].grid()
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5)) # insere um intervalo no eixo x de 5 e0 5 no caso
axs[0].yaxis.set_major_locator(plt.MultipleLocator(300)) # insere um intervalo no eixo y de 1000 em 1000 no caso


# subplot 2

axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Boxplot - Imigração do Brasil para o Canada')
axs[1].set_xlabel('Ano')
axs[1].set_ylabel('Imigrantes')
axs[1].grid()
```

# Comparativo dos maiores países da America Latina (imigração Canadá)

```python
fig, axs = plt.subplots(2,2,figsize=(18,6))

axs[0,0].plot(df.loc['Brasil',anos])
axs[0,0].set_title('Imigração do Brasil para o Canada \n 1980 a 2013')


axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('Imigração do Colômbia para o Canada \n 1980 a 2013')


axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Imigração do Argentina para o Canada \n 1980 a 2013')

axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Imigração do Peru para o Canada \n 1980 a 2013')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # insere um intervalo no eixo x de 5 e0 5 no caso

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Numero de imigrantes')


fig.subplots_adjust(hspace=0.5,wspace=0.3)
```
