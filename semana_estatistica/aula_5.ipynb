```python
top_df=df.sort_values("Total", ascending=False).head(10)
top_df
```

```python
import matplotlib.pyplot as plt
```

# Temas do seaborn

```python
ax= sns.barplot(data=top_df, y=top_df.index, x="Total", orient='h')
ax.set(title="Top 10 países com mais imigrantes para o Canadá",
             xlabel='Total de imigrantes',
             ylabel=''
)
plt.show()
```

# Gerando paleta de cores

### Função para gerar palleta de cores

```python
def gerar_grafico_paleta (palleta):
  flg, ax= plt.subplots( figsize=(14,8))
  ax=sns.barplot(data=top_df, y=top_df.index, x="Total", orient='h', palette=palleta)
  ax.set_title("Top 10 países com mais imigrantes para o Canadá",
              loc='left',
              fontsize=16)
  ax.set_xlabel('Total de imigrantes',fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

```python
gerar_grafico_paleta("viridis")
```

```python
gerar_grafico_paleta("colorblind")
```

```python
# ... código anterior ...

flg, ax= plt.subplots( figsize=(14,8))
ax = sns.barplot(data=top_df, y=top_df.index, x="Total", orient='h', palette='colorblind')
ax.set_title("Top 10 países com mais imigrantes para o Canadá", loc='left', fontsize=16)
ax.set_xlabel('Total de imigrantes',fontsize=14)
ax.set_ylabel('')

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```
