### Gerando o grafico

### Importando as bibliotecas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Gendo um histograma dos dados brutos

```python
pesquisa_musical.plot(kind='hist')
```

```python
pesquisa_musical.columns
```

### Gera um histograma por estilo de musica

```python


# Exemplo de DataFrame


df = pd.DataFrame(pesquisa_musical)

# Criar uma figura com múltiplos subplots
fig, axes = plt.subplots(nrows=len(df.columns), ncols=1, figsize=(10, len(df.columns) * 4))

# Se houver apenas uma coluna, axes não será uma lista de subplots
if len(df.columns) == 1:
    axes = [axes]

# Iterar sobre as colunas e criar histogramas
for col, ax in zip(df.columns, axes):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f'Histograma de {col}')

# Ajustar layout para que não haja sobreposição
plt.tight_layout()
plt.show()
```

### Soma todas as notas

```python
sum=pesquisa_musical.sum().sort_values(ascending=False)
sum
```

### Gera um grafico barra com a soma das notas

```python
sum.plot(kind='bar')
```

### Deixa o grafico na horizontal

```python
sum.plot(kind='barh')
```

```python
sns.histplot(sum,kde=True)
```

### Agora gera a media das notas e gera os graficos

```python
media=pesquisa_musical.mean().sort_values(ascending=False)
media
```

```python
media.plot(kind='barh')
```

```python


# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.barh(media.index, media.values, color='skyblue', edgecolor='black')

# Adicionar título e rótulos
plt.title('Media de Votos por Música')
plt.xlabel('media')
plt.ylabel('Musica')

# Rotacionar os rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=45)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
