# Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Carregando o DataSet

```python
url='dataset/pesquisa_musical_prof.csv'
df=pd.read_csv(url)
df.head()
```

# Renomeando das colunas do DataSet

```python
df.columns=['data','rock','rap','mpb','reggae','funk','sertanejo','gospel','eletronica','professor','faixa_etaria','samba','forro','trap','chave']
df.head()
```

# Removendo a coluna data do DataFrame

```python
del df['data']
df.head()
```

# Gerando um grafico barra encima da media de todas notas

```python
df_media=df[['rock','rap','mpb','reggae','funk','sertanejo','gospel','eletronica','samba','forro','trap']]
df_media.mean().sort_values(ascending=True).plot(kind='barh')
```

# Filtrando as notas e buscando somente as notas do Professor Damarcones

```python
df_professor = df[df['professor'] == 'Damarcones']
df_professor.head()
```

### Gerando o grafico encima da media das notas do professor Damarcones

```python
media=df_professor[['rock','rap','mpb','reggae','funk','sertanejo','gospel','eletronica','samba','forro','trap']].mean().sort_values(ascending=True)
media.plot(kind='barh')
```

### Detalhando o grafico

```python

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.barh(media.index, media.values, color='skyblue', edgecolor='black')

# Adicionar título e rótulos
plt.title('Media de Votos por Música')
plt.xlabel('media')
plt.ylabel('Musica')

# Rotacionar os rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=45)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

### Brincando com as cores das barras

```python
# Definir uma lista de cores para cada barra
cores = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightblue', 
         'lightpink', 'lightyellow', 'lightgray', 'lightgoldenrodyellow', 
         'lightcyan', 'lightsteelblue']

# Criar o gráfico de barras horizontais
plt.figure(figsize=(12, 8))
plt.barh(media.index, media.values, color=cores, edgecolor='black')

# Adicionar título e rótulos
plt.title('Média de Votos por Gênero Musical')
plt.xlabel('Média')
plt.ylabel('Gênero Musical')

# Ajustar o layout e mostrar o gráfico
plt.tight_layout()
plt.show()
```

# Inserindo "Romulo Silvestre" para as linhas nullas da coluna professor

```python
df_geral['professor'] = df_geral['professor'].fillna('Romulo Silvestre')
df_geral.head(100)
```

```python
df_geral.columns
```

# Gerando um histograma de cada professor e estilos de musicas

```python

df = pd.DataFrame(df_geral)

# Lista de estilos musicais
estilos = ['rock', 'rap', 'mpb', 'reggae', 'funk', 'sertanejo', 'gospel', 'eletronica', 'samba', 'forro', 'trap']

# Obter lista de professores únicos
professores = df['professor'].unique()

# Configurar o tamanho da figura
plt.figure(figsize=(15, 10))

# Criar histogramas para cada professor
for professor in professores:
    # Filtrar dados para o professor atual
    df_professor = df[df['professor'] == professor]
    
    # Configurar subplots para cada estilo musical
    plt.figure(figsize=(15, 10))
    plt.suptitle(f'Notas dos Estilos Musicais - Professor: {professor}', fontsize=16)
    
    for i, estilo in enumerate(estilos, start=1):
        plt.subplot(4, 3, i)  # Ajuste o layout conforme necessário
        plt.hist(df_professor[estilo].dropna(), bins=10, color='skyblue', edgecolor='black')
        plt.title(estilo)
        plt.xlabel('Nota')
        plt.ylabel('Frequência')
    
    # Ajustar o layout do gráfico
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```
